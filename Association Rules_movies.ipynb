```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
import seaborn as sns
```

```python
data = pd.read_csv('P:/python/my_movies.csv')
data
```

```python
data1 = data.iloc[:,5:]
data1
```

# Apriori Algoritham
###### support 10% and min_support = 0.1

```python
frequent_itemsets = apriori(data1, min_support=0.1, use_colnames=True)
frequent_itemsets
```

###### By giving support is 20% means min_support = 0.2

```python
frequent_itemsets1 = apriori(data1, min_support=0.2, use_colnames=True)
frequent_itemsets1
```

# Association rules

###### By giving metric as "confidence" and 70% confidence 

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
rules
```

###### By giving metric is "lift".

```python
rules2 = association_rules(frequent_itemsets, metric="lift", min_threshold=0.9)
rules2
```

```python
rules2.sort_values('lift',ascending = False)[0:20]
```

```python
rules2[rules2.lift>1]
```
